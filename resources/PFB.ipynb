```python
import os

avro_file_path = os.getenv("AVRO_FILE_PATH")

if avro_file_path is None:
    raise EnvironmentError("Environment variable 'AVRO_FILE_PATH' is not set.")
```

```python
from pfb.reader import PFBReader
import pandas as pd


try:
    with PFBReader(avro_file_path) as reader:

        # print("PFB Schema:")
        # print(reader.schema)

        # print("PFB Metadata:")
        # print(reader.metadata)

        records = list(reader)

        df = pd.DataFrame(records)


except Exception as e:
    print(f"Error: {e}")
```
